```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (12,5)
```

```python
from sklearn.manifold import TSNE
```

```python
from sklearn.datasets import load_digits
```

```python
digit = load_digits()
```

```python
img = digit.images
```

```python
img.shape
```

```python
plt.imshow(img[0], cmap=plt.cm.Greys_r)
```

```python
X = img.reshape(-1,64)
```

```python
X.shape
```

```python
tsne = TSNE(n_components=2, perplexity=30.)
```

```python
tsne.fit(X)
```

```python
Z = tsne.embedding_
```

```python
Z.shape
```

```python
plt.scatter(Z[:, 0], Z[:, 1])
```

```python
y = digit.target
```

```python
plt.scatter(Z[:, 0], Z[:, 1], c=y)
```

```python
permplex = [1, 3, 5, 10, 30, 50]
```

```python
for p in permplex:
    tsne = TSNE(n_components=2, perplexity=p)
    tsne.fit(X)
    Z = tsne.embedding_
    fit, ax = plt.subplots(1, 1)
    ax.scatter(Z[:,0], Z[:,1], c=y, cmap=plt.cm.seismic_r)
    ax.set_title('Perplexity {}'.format(p))
```
